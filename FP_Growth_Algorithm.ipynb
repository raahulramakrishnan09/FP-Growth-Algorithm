```python
import numpy as  np
import pandas as pd
```

```python
df=[['Milk','Onion','Nutmeg','Beans','Eggs','Yogurt'],
    ['Dill','Onion','Nutmeg','Beans','Eggs','Yogurt'],
    ['Milk','Apple','Beans','Eggs'],
    ['Milk','Unicorn','Corn','Beans','Yogurt'],
    ['Corn','Onion','Onion','Beans','Ice Cream','Eggs']]
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
#List_to_Dataframe
te=TransactionEncoder()
te_ary=te.fit_transform(df)
df=pd.DataFrame(te_ary,columns=te.columns_)
df
```

```python
#FP_Growth_Algorithm
from mlxtend.frequent_patterns import fpgrowth

fpgrowth(df,min_support=0.5)
```

```python
#providing_item_names
from mlxtend.frequent_patterns import fpgrowth

fpgrowth(df,min_support=0.3,use_colnames=True)
```
